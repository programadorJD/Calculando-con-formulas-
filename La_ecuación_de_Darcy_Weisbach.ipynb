# **Ecuacion de Darcy-Weisbach**

# Introducción:

La ecuación de Darcy-Weisbach describe cómo la presión de un fluido disminuye al fluir a través de una tubería debido a la fricción. Veamos qué significa cada parte de la ecuación:


ΔP (pérdida de presión): Es la disminución de presión que sufre el fluido al pasar por la tubería.


f (coeficiente de fricción): Es un valor que depende de la rugosidad de la tubería y del número de Reynolds (Re), que relaciona la velocidad del fluido con su viscosidad. Cuanto mayor sea f, mayor será la fricción y, por lo tanto, mayor será la pérdida de presión.


(L/D) (relación longitud-diámetro): Es la proporción entre la longitud de la tubería (L) y su diámetro interior (D). Cuanto mayor sea esta relación, mayor será la pérdida de presión.


(ρ * v^2 / 2) (termino de energía cinética): Representa la energía cinética del fluido, que depende de su densidad (ρ) y velocidad (v). Cuanto mayor sea la velocidad, mayor será la pérdida de presión.

# Conclusión:

En resumen, la ecuación dice que la pérdida de presión en una tubería es directamente proporcional a:


- La fricción (f)
- La longitud de la tubería (L)
- La velocidad del fluido (v)
- La densidad del fluido (ρ)


Y es inversamente proporcional al diámetro de la tubería (D). Esto significa que:


- Aumentar el diámetro de la tubería reduce la pérdida de presión.
- Aumentar la velocidad del fluido aumenta la pérdida de presión.
- Aumentar la longitud de la tubería aumenta la pérdida de presión.
- Aumentar la rugosidad de la tubería aumenta la pérdida de presión.
